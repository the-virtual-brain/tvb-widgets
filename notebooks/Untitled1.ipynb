```python
from tvb.datatypes.connectivity import Connectivity
from tvbwidgets.ui.connectivity_react.connectvity_widget_new import ConnectivityWidget

conn = Connectivity.from_file()
conn.configure()

w = ConnectivityWidget(conn)
display(w)
```

```python
import numpy as np

w = conn.weights
t = conn.tract_lengths

# Only look at connected pairs
mask = w != 0
connected_tracts = t[mask]

print("Connected tract lengths (mm):")
print(f"  Min:    {connected_tracts.min():.2f}")
print(f"  Max:    {connected_tracts.max():.2f}")
print(f"  Mean:   {connected_tracts.mean():.2f}")
print(f"  Median: {np.median(connected_tracts):.2f}")

# Distribution
percentiles = [10, 25, 50, 75, 90]
for p in percentiles:
    print(f"  {p}th percentile: {np.percentile(connected_tracts, p):.2f} mm")
```

```python
import numpy as np

# Check tract_lengths matrix
tl = conn.tract_lengths
print("tract_lengths shape:", tl.shape)
print("Non-zero per region:", np.count_nonzero(tl, axis=1))
print("Zero entries:", np.sum(tl == 0))
print("Min value:", tl.min())
print("Max value:", tl.max())
print("Sample values (first 5x5):")
print(tl[:5, :5])
```

```python
# Verify this theory
weights_nonzero = conn.weights != 0
tracts_nonzero  = conn.tract_lengths != 0

print("Regions with weight connection but no tract:", np.sum(weights_nonzero & ~tracts_nonzero))
print("Regions with tract but no weight connection:", np.sum(tracts_nonzero & ~weights_nonzero))
```
